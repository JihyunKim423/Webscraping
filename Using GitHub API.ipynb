##### 4. Write code that creates an SQL database + table, and stores all the information obtained in (3) in it.  Please be cautious of the data type you choose for each column and briefly justify your decisions.  What do you choose as Primary Key and why?

```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect('git_contributors_db')
c = con.cursor()

# Drop table if the table already exists
c.execute('''
DROP TABLE IF EXISTS contributors100 
''')
con.commit()
# Create a new table, and set proper datatype depending on the values of each columns
# Here, id should be primary key as it's the only distinctive, non-changeable value identifying a user.
c.execute('''
CREATE TABLE IF NOT EXISTS contributors100 (
          id    INTEGER PRIMARY KEY,
          login TEXT,
          location  TEXT,
          email TEXT,
          hireable   TEXT,
          bio   TEXT,
          twitter   TEXT,
          public_repos  INTEGER,
          public_gists  INTEGER,
          followers INTEGER,
          following INTEGER,
          created_at    DATETIME
)
''')
con.commit()

# Insert data into the predefined table
for i_2 in json_dict2:
    c.execute('''
    INSERT INTO contributors100 (
              id, login, location, email, hireable, bio, twitter, public_repos, public_gists, followers, following, created_at
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        i_2.get('id'),
        i_2.get('login'),
        i_2.get('location'),
        i_2.get('email'),
        i_2.get('hireable'),
        i_2.get('bio'),
        i_2.get('twitter_username'),
        i_2.get('public_repos'),
        i_2.get('public_gists'),
        i_2.get('followers'),
        i_2.get('following'),
        i_2.get('created_at')
    ))
    
con.commit()
```

```python
# Print the entire DB
c.execute('SELECT * FROM contributors100')
rows = c.fetchall()

for r in rows:
    print(r)
```

##### 5. Optimize your code in (4) to allow for quick look-ups of "login", "location", and "hireable".  I.e., I would like, for example, to run the command  <<  SELECT * FROM table WHERE location = "Tokyo"  >>  fast.  What choices do you make and why?

**Answer to the question:**

 simply add indices to login, location, and hireable columns. Indexing improves the data retrieval operation speed on a database table at the cost of additional writes and storage space to maintain the index. Queries that filter on an indexed column can use the index to quickly locate the data without scanning every row in the table, which becomes increasingly efficient as the size of the table grows.

```python
# Create indices on the login, location, and hireable column
c.execute('CREATE INDEX IF NOT EXISTS idx_login ON contributors100 (login)')
c.execute('CREATE INDEX IF NOT EXISTS idx_location ON contributors100 (location)')
c.execute('CREATE INDEX IF NOT EXISTS idx_hireable ON contributors100 (hireable)')
con.commit()
```

```python
c.execute('SELECT * FROM contributors100 WHERE location = "Tokyo, Japan"')
rows = c.fetchall()

for r in rows:
    print(r)
```
